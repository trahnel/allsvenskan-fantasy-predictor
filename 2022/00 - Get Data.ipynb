```python
import requests
import json
import pandas as pd
```

## Get basic data about players and gameweeks

```python
response = requests.get('https://fantasy.allsvenskan.se/api/bootstrap-static/')
```

```python
with open('data/during-season/bootstrap-static.json', 'w+') as json_file:
    json.dump(response.json(), json_file, indent=2)
```

## Get all players detailed data

```python
data = response.json();
player_data = pd.DataFrame(data['elements'])
player_data.sample(5)
```

```python
player_data.shape[0]
```

```python
from IPython.display import clear_output

df_height = player_data.shape[0]

for index, player in player_data.iterrows():
    player_id = str(player['id'])
    first_name = player['first_name']
    last_name = player['second_name']
    
    
    clear_output(wait=True)
    print(f'Getting player {player_id} {first_name} {last_name} (Total: {df_height})')

    player_response = requests.get(f'https://fantasy.allsvenskan.se/api/element-summary/{player_id}/')
    
    file_name = f"{player_id.rjust(3, '0')}_{first_name}_{last_name}".replace(' ', '_')
    with open(f'data/during-season/players/{file_name}.json', 'w+') as player_file:
        json.dump(player_response.json(), player_file, indent=2)
        
clear_output(wait=True)
print('Done!')
```
