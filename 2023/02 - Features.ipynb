```python
import json
import pandas as pd
import os
```

```python
def flatten(xss):
    return [x for xs in xss for x in xs]
```

```python
# Load all matches into a data frame
dir = 'data/matches'
all_players = pd.DataFrame()

for file_name in os.listdir(dir):
    with open(os.path.join(dir, file_name), 'r') as json_file:
        match = json.loads(json_file.read())
    
    teams = match['content']['lineup']['lineup']
    
    for team in teams:
        players = pd.json_normalize(flatten(team['players']) + team['bench'])
        all_players = pd.concat([all_players, players], ignore_index=True)
```
